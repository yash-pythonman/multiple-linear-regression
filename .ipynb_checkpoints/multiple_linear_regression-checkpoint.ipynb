```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

winequality_data=pd.read_csv("dataset/winequality.csv")
Y=winequality_data['quality']
X = winequality_data[winequality_data.columns[0:11]]
X_train, x_test, Y_train, y_test=train_test_split(X,Y)
multiple_linear_regressor=LinearRegression()
multiple_linear_regressor.fit(X_train, Y_train)
y_predict = multiple_linear_regressor.predict(x_test)

for column in X.columns:
    plt.scatter(X_train[column], Y_train)
    plt.title(f"Plot between Wine 'quality' and '{column}'")
    plt.scatter(x_test[column],y_predict, color="black", marker="+")
    plt.scatter(x_test[column],y_predict, color="r", marker="*")
    plt.legend(labels=['Trained_Response','Actual_quality', "Predicted_quality"])
    plt.xlabel(column)
    plt.ylabel("quality")
    plt.show()

for column, coefficent in zip(X.columns, multiple_linear_regressor.coef_):
    print(f"coefficent with {column} = ", coefficent )
print("Intercept  = ", multiple_linear_regressor.intercept_)
```
